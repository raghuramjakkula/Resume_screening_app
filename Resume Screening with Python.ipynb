# Cleaning Data:                                      
1 URLs,                                                
2 hashtags,                                             
3 mentions,                                                     
4 special letters,                                             
5 punctuations:                                             

```python
import re                       ## re regular expression using *,@,# special characters remove
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))   ## lambda function using clean resume skills
```

```python
df['Resume'][0]
```

```python
from sklearn.preprocessing import LabelEncoder    ## labelencoder using catgorical data into numerical data
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

# Vactorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer  ## tfidf - term frequency & inverse document frequency
tfidf = TfidfVectorizer(stop_words='english')
                                                           ## remove ing the stop words
tfidf.fit(df['Resume'])
requredTaxt  = tfidf.transform(df['Resume'])
```

```python
from sklearn.model_selection import train_test_split     ## spliting the data traing , testing
```

```python
X_train, X_test, y_train, y_test = train_test_split(requredTaxt, df['Category'], test_size=0.2, random_state=42)   ##test size 20%
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier   # using ml model kneighbors classification 
from sklearn.multiclass import OneVsRestClassifier   # one vs all classication n
from sklearn.metrics import accuracy_score           # accuracy scorer

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)                            ## fit the model
ypred = clf.predict(X_test)
print(accuracy_score(y_test,ypred))                 ## accuracy 98%
```
